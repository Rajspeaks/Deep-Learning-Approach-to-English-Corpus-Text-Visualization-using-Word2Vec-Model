# **English Corpus Text-Visualization using Word2Vec Model**

```python
import gensim
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot
import matplotlib.font_manager as fm
```

```python
t1=input("Enter Text:")
```

```python
import re
whitespace = re.compile(u"[\s\u0020\u00a0\u1680\u180e\u202f\u205f\u3000\u2000-\u200a]+", re.UNICODE)
english_fullstop = u"\u0964"
punctSeq   = u"['\"“”‘’]+|[.?!,…]+|[:;]+"
punc = u"[(),$%^&*+={}\[\]:\"|\'\~`<>/,€¦!?½£¶™¼©⅐⅑⅒⅓⅔⅕⅖⅗⅘⅙⅚⅛⅜⅝⅞⅟↉¤¿º;-]+"
t1= whitespace.sub(" ",t1).strip()
t1 = re.sub(punctSeq, " ", t1)
t1 = re.sub(english_fullstop, " ",t1)
t1 = re.sub(punc, " ", t1)
```

```python
print(t1)
```

```python
doc=[t1.split()]
```

```python
model = Word2Vec(doc, min_count=1)
```

```python
model.train(doc, total_examples=len(doc), epochs=model.iter)
```

```python
x= model[model.wv.vocab]
pca = PCA(n_components=2)
result = pca.fit_transform(x)
```

```python
pyplot.scatter(result[:, 0], result[:, 1])
words = list(model.wv.vocab)
prop = fm.FontProperties(fname='Melon Honey.ttf')
for i, word in enumerate(words):
	pyplot.annotate(word, xy=(result[i, 0], result[i, 1]),fontproperties=prop)
pyplot.show()
```
